# Binary tree

```python
class Node:
    # constructor
    def __init__(self, data = None, right = None, left = None):
        self.data = data
        self.right = right
        self.left = left

class BinarySearchTree:
    def __init__(self, data = None):
        if data:
            self.root = Node(data)
        else:
            self.root = None
    
    def insert(self, data):
        if not self.root:
            self.root = Node(data)
            return data
        current_node = self.root
        while current_node:
            if data == current_node.data:
                return
            if data < current_node.data:
                if not current_node.left:
                    current_node.left = Node(data)
                    return data
                current_node = current_node.left
            else:
                if not current_node.right:
                    current_node.right = Node(data)
                    return data
                current_node = current_node.right

    def delete(self, data):
        currentNode = self.root
        if currentNode.data == data and not (currentNode.right or currentNode.left):
            self.root = None
            return
        while currentNode:
            if currentNode.data > data:
                prevNode = currentNode
                currentNode = currentNode.left
            elif currentNode.data < data:
                prevNode = currentNode
                currentNode = currentNode.right
            else:
                if currentNode.right:
                    minNode = currentNode.right
                    while minNode.left:
                        prevNode = minNode
                        minNode = minNode.left
                    currentNode.data = minNode.data
                    prevNode.left = minNode.right
                elif currentNode.left:
                    maxNode = currentNode.left
                    while maxNode.right:
                        prevNode = maxNode
                        maxNode = maxNode.right
                    currentNode.data = maxNode.data
                    prevNode.right = maxNode.left
                else:
                    if data > prevNode.data:
                        prevNode.right = None
                    else: 
                        prevNode.left = None
                return data




    def print(self):
        self.printTree(self.root)

    def printTree(self, node, level=0):
        if node != None:
            self.printTree(node.right, level + 1)
            print(' ' * 4 * level + '-> ' + str(node.data))
            self.printTree(node.left, level + 1)



tree = BinarySearchTree()
tree.insert(20)
tree.insert(25)
tree.insert(32)
tree.insert(14)
tree.insert(7)
tree.insert(18)
tree.insert(9)

tree.print()

print('-----')
tree.delete(18)
tree.print()
```
